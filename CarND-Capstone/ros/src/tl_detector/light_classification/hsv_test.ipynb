```python
import cv2
import numpy as np 
from IPython.display import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = './data/red.png'
Image(image)
```

```python
img = mpimg.imread('./data/red.png')

image_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
```

```python
H = image_hsv[:,:,0]
S = image_hsv[:,:,1]
V = image_hsv[:,:,2]


plt.figure()
plt.imshow(H)
plt.figure()
plt.imshow(S)
plt.figure()
plt.imshow(V)
```

```python
img = mpimg.imread('./data/redLight.png')
image_hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(image_hsv)
```

```python
H = image_hsv[:,:,0]
S = image_hsv[:,:,1]
V = image_hsv[:,:,2]

plt.figure()
plt.imshow(H)
plt.figure()
plt.imshow(S)
plt.figure()
plt.imshow(V)
```

```python
binary_output = np.zeros_like(H)
binary_output[(H>0) & (H<=10)] = 1
# binary_output[(V>0.8) & (V<=1.0)] = 1
plt.imshow(binary_output, cmap='gray')
```
